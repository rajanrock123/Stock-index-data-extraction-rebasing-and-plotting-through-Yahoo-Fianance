# Importing Relevant Python Packages 

```python
import numpy as np
import pandas as pd
import yfinance as yf 
import matplotlib.pyplot as plt
```

# Downloading Historical Prices for multiple Stocks

```python
data= yf.download("SPY AAPL", start="2021-01-01", end="2022-09-23")
```

# Rebasing the Stock Prices to 100 for Comparison of Performance 

```python
data['Rebased_AAPL'] = (100/(data['Close']['AAPL'][0]))*(data['Close']['AAPL'])
data['Rebased_SPY'] = (100/(data['Close']['SPY'][0]))*(data['Close']['SPY'])
```

```python
data.head(5)
```

# Plotting the Line Charts for the rebased Hstorical Stock Prices to gauge their performance 

```python
fig, ax = plt.subplots(figsize=(25, 10))
ax.plot(data.index,data['Rebased_AAPL'], 'r', label='AAPL Rebased') 
ax.plot(data.index,data['Rebased_SPY'], '#008000', label='SPY Rebased')
plt.rcParams.update({'font.size':20})
```
